```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot data
# Series
data = pd.Series(np.random.randn(1000), 
                 index=np.arange(1000))
```

```python
# 对生成的数据进行累加
data = data.cumsum()
```

```python
data.plot()
```

```python
# 显示出来即可
plt.show()
```

```python
# DataFrame
data = pd.DataFrame(np.random.randn(1000, 4),
                     index=np.arange(1000),
                     columns=list("ABCD"))
```

```python
data = data.cumsum()
```

```python
data.head()
```

```python
data.plot()
```

```python
plt.show()
```

```python
# plot method:
# 'bar', 'hist', 'box', 'kde', 'area', 'scatter', 
# 'hexbin', 'pie'
data.plot.scatter(x='A', y='B')
```

```python
plt.show()
```

```python
ax = data.plot.scatter(x='A', y='B', color='DarkBlue',
                       label='Class1')
```

```python
data.plot.scatter(x='A', y='C', color='DarkGreen',
                  label='Class2', ax=ax)
```

```python
plt.show()
```
